## Agg CarNumber

```python
def group_by_values_series(ser):
    return ser.groupby(ser).agg(lambda x: list(x.index)).sort_index(ascending=False)
```

```python
df.reset_index(inplace=True)
```

```python
agg_count_fines = df.CarNumber.value_counts()
agg_count_fines
```

```python
#or
group_by_values_series(agg_count_fines).rename_axis('CountFines')
```

```python
df[df.CarNumber == df.CarNumber.iloc[0]]
```

```python
agg_sum_fines = df.groupby('CarNumber').Fines.sum().sort_values()
agg_sum_fines
```

```python
#or
group_by_values_series(agg_sum_fines).rename_axis('SumFines')
```

```python
df[df.CarNumber == df.CarNumber.iloc[0]]
```
